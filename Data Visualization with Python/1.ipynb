```python
df_tot = pd.DataFrame(df[years].sum(axis=0))
df_tot.index = list(map(int,df_tot.index))
df_tot.reset_index(inplace=True)
df_tot.columns = ['Years','Total']
df_tot
```

```python
df_tot.plot(kind='scatter',x='Years',y='Total',figsize=(14,8),color='darkblue')
plt.title('Scatter plot for all years')
plt.xlabel('Years')
plt.ylabel('Total No. of immigrants')
plt.show()
```

```python
x = df_tot['Years']
y = df_tot['Total']
fit = np.polyfit(x,y,deg=2)
fit
```

```python
df_tot.plot(kind='scatter',x = 'Years',y = 'Total',figsize=(10,6),color='darkblue')
plt.title('Total No. of immigrants')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')
plt.plot(x,x*x*fit[0]+x*fit[1]+fit[2],color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()
print('No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]))
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('alice_novel.txt','r').read()
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(
    background_color = 'black',
    max_words = 2000,
    stopwords=stopwords
)
alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')
alice_wc.generate(alice_novel)
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

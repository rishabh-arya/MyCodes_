## Week - 2

nx.clustering -->  LCC (Local Clustering Coefiicient)

nx.average_clustering -->  Global Clustering Coefiicient (Approach - 1)

nx.transitivity --> Global Clustering Coefficient (Approach - 2)

nx.shortest_path(G,'node1','node2') --> list of nodes within shortest_path

nx.shortest_path_length(G,'node1','node2') --> length of shortest_path

nx.average_shortest_path_length(G) --> Average_shortest_path_length of the graph

nx.shortest_path_length(G,'node') --> Dictionary of shortest path length from node n to others

nx.bfs_tree(G,'src_node') --> BFS Traversal from the source node

nx.eccentricity(G) --> Dictionary of nodes with values as largest distance between node n to others

nx.diameter(G) --> Diameter of the Graph

nx.radius(G) --> Radius of the Graph

nx.center(G) --> list of nodes having eccentriticity equal to radius

nx.periphery(G) --> list of nodes having eccentriticity equal to diameter

nx.is_connected(G) --> Bool, Tells us whether the undirected graph is connected or not

nx.connected_components(G) --> list of sets of connected components of Graph

nx.number_connected_components(G) --> No. of connected components

nx.node_connected_component(G,'node') --> Set of connected component nodes

nx.is_strongly_connected(G) --> Tells whether a directed graph is strongly connected or not

nx.strongly_connected_components(G) --> list of set of strongly connected components
scc= nx.strongly_connected_components(G)
max(scc,key=len) --> will return the length or number of nodes in the largest connected component

nx.strongly_connected_component_subgraphs(G) --> returns sub graphs of the connected components
sub_graphs= nx.strongly_connected_components(G)
max(sub_graphs,key=len) --> returns sub graph of the largest connected component

nx.is_weakly_connected(G) --> Tells whether the graph is weakly connected or not

nx.weakly_connected_components(G) --> list of set of weakly connected components

nx.node_connectivity(G,'node1','node2') --> Minimum number of nodes to disconnect node1 and node2

nx.minimum_node_cut(G) -->  Set of Minimum number of nodes to disconnect node1 and node2

nx.edge_connectivity(G) --> Smallest number of edges that can be removed from this graph in order to                                    disconnect it
nx.minimum_edge_cut(G) --> Set of tuples or edges to remove to disconnect the graph

nx.all_simple_paths(G,'node1','node2') --> List of list of all possible paths from node1 to node2

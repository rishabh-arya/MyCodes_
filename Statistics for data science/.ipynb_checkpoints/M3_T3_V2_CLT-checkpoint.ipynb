```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
import random
```

### Create a normally distributed population

```python
np.random.seed(1)
norm_pop = np.random.normal(10,2,10000)
sns.kdeplot(norm_pop)
```

### Create a sampling distribution by sampling the normal distribution population

```python
# 1. create random samples
# 2. calculate the mean values for each of these samples
# 3. analyze the mean distribution
```

```python
def sampling_distr(pop, n):
    sample_mean = []
    for i in np.arange(0,100,1):
        np.random.seed(1)
        sample_mean.append(np.array(random.sample(list(pop),n)).mean())
    return sample_mean
```

```python
sampling_distr(norm_pop,10)
```

```python
sns.kdeplot(norm_pop)
sns.kdeplot(sampling_distr(norm_pop,10))
sns.kdeplot(sampling_distr(norm_pop,50))
```

```python
norm_pop.mean(), np.array(sampling_distr(norm_pop,10)).mean(),  np.array(sampling_distr(norm_pop,50)).mean()
```

```python
norm_pop.std(), np.array(sampling_distr(norm_pop,10)).std(),  np.array(sampling_distr(norm_pop,50)).std()
```

### Create a exponential distribution population

```python
np.random.seed(1)
exp_pop = np.random.exponential(size = 10000)
sns.kdeplot(exp_pop)
```

### Create a sampling distribution by sampling the normal distribution population

```python
sns.kdeplot(exp_pop)
sns.kdeplot(sampling_distr(exp_pop,10))
sns.kdeplot(sampling_distr(exp_pop,50))
```

```python
exp_pop.mean(), np.array(sampling_distr(exp_pop,10)).mean(), np.array(sampling_distr(exp_pop,50)).mean()
```

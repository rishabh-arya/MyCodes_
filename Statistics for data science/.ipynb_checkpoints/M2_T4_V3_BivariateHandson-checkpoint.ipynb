# Analysis of Car Mileage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
cars = pd.read_csv('cars.csv')
cars
```

```python
# If the displacement is high then the mileage should be low
```

```python
sns.scatterplot(cars.disp, cars.mpg)
```

```python
# No of cyl. is related to mileage
```

```python
sns.scatterplot(cars.cyl, cars.mpg)
```

```python
sns.boxplot(cars.cyl, cars.mpg)
```

```python
# qsec and mpg are related
```

```python
sns.scatterplot(cars.qsec, cars.mpg)
```

```python
# HP and mpg are related
```

```python
sns.scatterplot(cars.hp, cars.mpg)
```

```python
# Gears and mpg
```

```python
sns.boxplot(cars.gear, cars.mpg)
```

```python
sns.pairplot(cars)
```

```python
cars.corr()
```

```python
plt.figure(figsize = (10, 8))
sns.heatmap(cars.corr(), annot = True, cmap = 'rainbow' )
```

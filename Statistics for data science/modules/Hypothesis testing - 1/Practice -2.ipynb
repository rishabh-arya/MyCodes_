```python
import numpy as np
import scipy.stats as stats
```

```python
# Sample of 25 Employees have average time spend of 62 minutes with 10 minutes of standard deviation
# Our hypothesised value is 60 mintues and we want that the mean of population to be greater than
# 60 minutes to have better improvements in learning
```

```python
# Step 1 -------------------------------------------------------------
N = 25
X = 62 # Sample mean
mean = 60 # Population mean
std = 10 # Sample std
# A right tail test 
# NULL Hypothesis --->   population mean <= 60
# Alternate Hypothesis ---> population mean > 60
```

```python
# Step 2 -------------------------------------------------------------
# Defining the acceptance criteria 
alpha = 0.05
# Since the number of instances is less than 30 we use T-Statistics
# Tcrit = 1.71
Tcrit = stats.t.ppf(q = 0.95,df=24)
```

```python
# Step 3-a ------------------------------------------------------------
# Calculating the Tstat value
# Formula is Tstat = (X - mean)/(std/ sqrt(N))
Tstat = (62 - mean)/(std/np.sqrt(N))
print(Tstat)
```

```python
# Step 3-b ------------------------------------------------------------
# Calculating the p-value
# pval = 0.164
pval = 1 - stats.t.cdf(x = 62,loc=60,scale = 10/(np.sqrt(N)),df=24)
pval
```

```python
# Step 4 -------------------------------------------------------------
# Decision
# Since pval is greater than alpha and the Tstat is also less than Tcrit we cannot reject NULL Hypothesis,
# i.e. we don't have enough evidence to reject NULL hypothesis and we'll go with NULL Hypothesis as truth
```

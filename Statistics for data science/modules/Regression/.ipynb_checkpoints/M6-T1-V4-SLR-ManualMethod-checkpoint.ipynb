# Braking Distance Estimation using Simple Linear Regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BrakingDistance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.scatter(df.speed, df.dist)
plt.title('Speed Vs Distance')
plt.xlabel('Speed')
plt.ylabel('Distance')
```

```python
df.corr()
```

```python
sns.distplot(df.speed)
```

```python
sns.distplot(df.dist)
```

```python
xbar = np.mean(df.speed)
xbar
```

```python
ybar = np.mean(df.dist)
ybar
```

```python
beta1 = np.sum((df.dist - ybar)*(df.speed - xbar))/np.sum((df.speed-xbar)**2)
beta1
```

```python
beta0 = ybar - beta1*xbar
beta0
```

```python
# y = -17.57 + 3.93 * x
df['pred'] = beta0 + beta1 * df.speed
df
```

```python
plt.scatter(df.speed, df.dist)
plt.plot(df.speed, df.pred, color = 'r')
plt.title('Speed Vs Distance')
plt.xlabel('Speed')
plt.ylabel('Distance')
```

```python
sns.kdeplot(df.pred)
sns.kdeplot(df.dist)
```

```python
df['resid'] = df.pred - df.dist
df
```

```python
sns.kdeplot(df.resid)
```

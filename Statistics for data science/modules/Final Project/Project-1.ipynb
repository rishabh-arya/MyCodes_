```python
from bioinfokit.analys import stat
res = stat()
res.tukey_hsd(df_new,res_var='Calories',
             xfac_var='Milk',
             anova_model = 'Calories ~ C(Milk)')
res.tukey_summary
```

```python
import statsmodels.formula.api as smf
formula = 'Calories ~  Category + Beverage + Size + Milk + TotalFat + TransFat + SaturatedFat  + Sodium + TotalCarb + Cholesterol + DietaryFibre + Sugars + Protein'
lm = smf.ols(formula = formula, data = df).fit()
print(lm.summary())
```

```python
formula = "Calories ~ TotalFat + Protein + Sugars"
model = smf.ols(formula = formula,data = df).fit()
print(model.summary())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_tmp = df[['Calories','Sugars','Protein','TotalFat']]
df_tmp = pd.DataFrame(scaler.fit_transform(df_tmp),columns=df_tmp.columns)
df_tmp
```

```python
formula = "Calories ~ TotalFat + Protein + Sugars"
model = smf.ols(formula = formula,data = df_tmp).fit()
print(model.summary())
```

```python
formula = "Calories ~ TransFat + Protein + Sugars"
model = smf.ols(formula = formula,data = df).fit()
print(model.summary())
```

```python
pred = model.predict(df)
```

```python
sns.scatterplot(pred,df.Calories)
```
